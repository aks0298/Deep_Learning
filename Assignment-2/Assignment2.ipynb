```python
import random
import numpy as np

def traintestsplit(X, y, testsize=0.3, randomseed=None):
    data = list(zip(X, y))

    if randomseed:
        random.seed(randomseed)

    random.shuffle(data)

    idx = int(len(data) * (1 - testsize))

    train = data[:idx]
    test = data[idx:]

    X_train, y_train = zip(*train)
    X_test, y_test = zip(*test)

    return np.array(X_train), np.array(X_test), np.array(y_train), np.array(y_test)
```

```python
X_train,X_test,y_train,y_test=traintestsplit(X,y,0.2,42)
```

# scaling the inputs
standard scaler-> z=(x-μ)/σ

```python
class StandardScaler:
  def __init__(self):
    self.mean=None
    self.std=None

  def fit(self,X):
    X=np.array(X)
    self.mean=X.mean(axis=0)
    self.std=X.std(axis=0)

  def transform(self,X):
    X=np.array(X)
    X_scaled=(X-self.mean)/self.std
    return X_scaled

  def fit_transform(self,X):
    X=np.array(X)
    self.fit(X)
    return self.transform(X)

  def inverse_transform(self,X_scaled):
    X_scaled=np.array(X_scaled)
    X=X_scaled*self.std+self.mean
    return X
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
def sigmoid(z):
  return 1/(1+np.exp(-z))
```

```python
def predict_prob(X,w,b):
  z=np.dot(X,w)+b
  p=sigmoid(z)
  return p
```

```python
def loss(y,p):
  return -np.mean(y*np.log(p)+(1-y)*np.log(1-p))
```

```python
def update_weights(X,y,w,b,lr):
  p=predict_prob(X, w, b)
  error=p-y
  dw=np.dot(X.T, error)/len(y)
  db=np.mean(error)
  w=w-lr*dw
  b=b-lr*db
  return w,b
```

```python
w=np.zeros(X_train.shape[1])
b=0.0
lr=0.1
epochs=200
```

```python
for i in range(epochs):
    w,b=update_weights(X_train_scaled,y_train,w,b,lr)
    print(f'epoch {i+1}: {loss(y_train,predict_prob(X_train_scaled,w,b))}')
```

```python
def predict_label(p,thr=0.5):
  return (p>=thr).astype(int)
```

```python
y_pred=predict_label(predict_prob(X_test_scaled,w,b),0.7)
```

```python
np.mean(y_pred==y_test)
```
