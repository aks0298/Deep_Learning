```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
#AND
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,0,0,1])
df=pd.DataFrame({'X1': x[:,0], 'X2': x[:,1], 'Y': y})
print(df)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#perceptron
model = Sequential()
model.add(Dense(1,input_dim=2,activation='sigmoid'))
```

```python

from tensorflow.keras.optimizers import Adam
model.compile(loss='binary_crossentropy',optimizer=Adam(learning_rate=0.1),metrics=['accuracy'])

model.fit(x,y,epochs=50)
print((model.predict(x)>0.5).astype(int))
```

```python
w,b=model.layers[0].get_weights()
print(w)
print(b)
```

```python
#OR
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,1,1,1])
df=pd.DataFrame({'X1': x[:,0], 'X2': x[:,1], 'Y': y})
print(df)

model=Sequential()
model.add(Dense(1,input_dim=2,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer=Adam(learning_rate=0.1),metrics=['accuracy'])
model.fit(x,y,epochs=100)
```

```python
model.layers[0].get_weights()
```

```python
x_test = np.array([[0,0],[0,1],[1,1],[0,0]])

print((model.predict(x_test) > 0.5).astype(int))
```

```python
#NOT
x=np.array([0,1],dtype=np.float32)
y=np.array([1,0],dtype=np.float32)

model=Sequential()
model.add(Dense(1,input_dim=1,activation='sigmoid'))
print(model.summary())
model.compile(loss='binary_crossentropy',optimizer=Adam(learning_rate=0.1),metrics=['accuracy'])
model.fit(x,y,epochs=100)

print(model.layers[0].get_weights())
x_test=np.array([0,0])
print((model.predict(x_test)>0.5).astype(int))
```

```python
#NAND
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([1,1,1,0])
df=pd.DataFrame({'X1': x[:,0], 'X2': x[:,1], 'Y': y})
print(df)

model=Sequential()
model.add(Dense(1,input_dim=2,activation='sigmoid'))
print(model.summary())
model.compile(loss='binary_crossentropy',optimizer=Adam(learning_rate=0.1),metrics=['accuracy'])
model.fit(x,y,epochs=70)

print(model.layers[0].get_weights())
print((model.predict(x)>0.5).astype(int))
```
